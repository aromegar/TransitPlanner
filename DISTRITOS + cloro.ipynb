## cargamos el json

```python
import json
geo_json = r"DATA/distritos.geojson"
```

```python
with open (geo_json) as geo_file:
    madrid = json.load(geo_file)
```

## cargamos el data:

```python
df = pd.read_csv("DATA/dist_barr_pobl.csv")
df
```

## agrupamos por distritos:

```python
code=[str(i) for i in range(1,22)]
```

```python
dfg = df.groupby(df.Distrito, sort = False).sum()
dfg["código"]=code
dfg
```

## Creamos el mapa y el coroplético:

```python
madrid_map = folium.Map(location=[40.4216,-3.6780], zoom_start=10, tiles="cartodbpositron");

madrid_map.choropleth(
    geo_data = madrid,
    data = dfg, 
    columns = ["código", "Población"],
    key_on = "feature.properties.coddistrit", 
    fill_color = "YlGnBu",
    fill_opacity = 0.4,
    line_opacity = 1,
    legend_name = "Población total por distritos",
    smooth_factor = 1,
    highlight=True
);
```

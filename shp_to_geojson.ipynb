# .shp   -->   .geojson

```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map, TileLayer
from folium.plugins import HeatMap, MarkerCluster
import geopandas as gpd
import pandas as pd
import json
```

## probamos los barrios:

```python
barrios_gdf = gpd.read_file("../Datos/Barrios Madrid/200001882.shp")
```

```python
barrios_gdf.crs = "EPSG:4258"
```

```python
barrios_gdf.head()
```

```python
madrid_map = folium.Map(location=[40.4216,-3.6780], zoom_start=12, tiles="cartodbpositron");
folium.GeoJson(data=barrios_gdf["geometry"]).add_to(madrid_map);
madrid_map
```
